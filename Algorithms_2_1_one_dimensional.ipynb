```python
"""Analysis of one dimension optimisation algorithms, Task_2_1, Bestuzhev, C4132"""

import math
import random
import matplotlib.pyplot as plt

precision=0.001    # common for all implemented algorithms precision


# Defined functions

def x_exp_3(x):
    return (x**3)
def abs_x(x):
    return(abs(x-0.2))
def sin_x(x):
    return(math.sin(x)/x)

# Simple-iteration methods of optimisation,
# return x: f(x)-->min, f(x_min) and counter of function calls.

def brute_force_x_exp_3 (x_min=0, x_max=1):
    step=(x_max-x_min)*precision
    x_opt=x_min
    x=x_min
    f_min=x_exp_3(x_min)
    f=f_min
    counter=1
    while x<x_max:
        x+=step
        f=x_exp_3(x)
        counter+=1
        if (f_min>f):
            f_min=f
            x_opt=x
    return({'x_opimal':x_opt, 'minimum':f_min, 'counter':counter})

def brute_force_abs_x (x_min=0, x_max=1):
    step=(x_max-x_min)*precision
    x_opt=x_min
    x=x_min
    f_min=abs_x(x_min)
    f=f_min
    counter=1
    while x<x_max:
        x+=step
        f=abs_x(x)
        counter+=1
        if (f_min>f):
            f_min=f
            x_opt=x
    return({'x_opimal':x_opt, 'minimum':f_min, 'counter':counter})

def brute_force_sin_x (x_min=0.1, x_max=1):
    step=(x_max-x_min)*precision
    x_opt=x_min
    x=x_min
    f_min=sin_x(x_min)
    f=f_min
    counter=1
    while x<x_max:
        x+=step
        f=sin_x(x)
        counter+=1
        if (f_min>f):
            f_min=f
            x_opt=x
    return({'x_opimal':x_opt, 'minimum':f_min, 'counter':counter})

# Dichotomy methods of optimisation,
# return x: f(x)-->min, f(x_min) and counter of function calls.

def dichotomy_x_exp_3(x_min=0, x_max=1):
    shift=precision/2
    x_opt=x_min
    f_opt=x_exp_3(x_min)
    counter=1
    while (abs(x_max-x_min)>precision):
        x_vice_min=(x_max+x_min)/2-shift
        x_vice_max=(x_max+x_min)/2+shift
        f_vice_min=x_exp_3(x_vice_min)
        f_vice_max=x_exp_3(x_vice_max)
        counter+=2
        if (f_vice_min<f_vice_max):
            x_max=x_vice_max
        elif (f_vice_min>f_vice_max):
            x_min=x_vice_min
        else:
            x_min=(x_max+x_min)/2
        if f_opt>min(f_vice_max, f_vice_min) and (x_max-x_min<=precision):
            if (f_vice_min<=f_vice_max):
                f_opt=f_vice_min
                x_opt=x_vice_min
            else:
                f_opt=f_vice_max
                x_opt=x_vice_max  
    return({'x_opimal':x_opt, 'minimum':f_opt, 'counter':counter})

def dichotomy_abs_x(x_min=0, x_max=1):
    shift=precision/2
    x_opt=x_min
    f_opt=abs_x(x_min)
    counter=1
    while (abs(x_max-x_min)>precision):
        x_vice_min=(x_max+x_min)/2-shift
        x_vice_max=(x_max+x_min)/2+shift
        f_vice_min=abs_x(x_vice_min)
        f_vice_max=abs_x(x_vice_max)
        counter+=2  
        if (f_vice_min<f_vice_max):
            x_max=x_vice_max
        elif (f_vice_min>f_vice_max):
            x_min=x_vice_min
        else:
            x_min=(x_max+x_min)/2
        if f_opt>min(f_vice_max, f_vice_min) and (x_max-x_min<=precision):
            if (f_vice_min<=f_vice_max):
                f_opt=f_vice_min
                x_opt=x_vice_min
            else:
                f_opt=f_vice_max
                x_opt=x_vice_max 
    return({'x_opimal':x_opt, 'minimum':f_opt, 'counter':counter})

def dichotomy_sin_x(x_min=0, x_max=1):
    shift=precision/2
    x_opt=x_min
    f_opt=sin_x(x_min)
    counter=1
    while (abs(x_max-x_min)>precision):
        x_vice_min=(x_max+x_min)/2-shift
        x_vice_max=(x_max+x_min)/2+shift
        f_vice_min=sin_x(x_vice_min)
        f_vice_max=sin_x(x_vice_max)
        counter+=2  
        if (f_vice_min<f_vice_max):
            x_max=x_vice_max
        elif (f_vice_min>f_vice_max):
            x_min=x_vice_min
        else:
            x_min=(x_max+x_min)/2
        if f_opt>min(f_vice_max, f_vice_min) and (x_max-x_min<=precision):
            if (f_vice_min<=f_vice_max):
                f_opt=f_vice_min
                x_opt=x_vice_min
            else:
                f_opt=f_vice_max
                x_opt=x_vice_max 
    return({'x_opimal':x_opt, 'minimum':f_opt, 'counter':counter})

# GOld-ratio methods of optimisation,
# return x: f(x)-->min, f(x_min) and counter of function calls.

def goldratio_x_exp_3(x_min=0, x_max=1):
    x_1=x_min+((3-math.sqrt(5))/2)*(x_max-x_min)
    x_2=x_max+((math.sqrt(5)-3)/2)*(x_max-x_min)
    f_x_1=x_exp_3(x_1)
    f_x_2=x_exp_3(x_2)
    counter=2
    x_opt=x_1
    f_opt=f_x_1
    while (abs(x_max-x_min)>precision):
        if (f_x_1<=f_x_2):
            x_max=x_2
            x_1,x_2=x_min+((3-math.sqrt(5))/2)*(x_max-x_min), x_1
            f_x_1,f_x_2=x_exp_3(x_1),f_x_1
            counter+=1
            x_opt=x_1
            f_opt=f_x_1
        else:
            x_min=x_1
            x_1,x_2=x_2,x_max+((math.sqrt(5)-3)/2)*(x_max-x_min)
            f_x_1,f_x_2=f_x_2,x_exp_3(x_2)
            counter+=1
            x_opt=x_2
            f_opt=f_x_2
    return({'x_opimal':x_opt, 'minimum':f_opt, 'counter':counter})

def goldratio_abs_x(x_min=0, x_max=1):
    x_1=x_min+((3-math.sqrt(5))/2)*(x_max-x_min)
    x_2=x_max+((math.sqrt(5)-3)/2)*(x_max-x_min)
    f_x_1=abs_x(x_1)
    f_x_2=abs_x(x_2)
    counter=2
    x_opt=x_1
    f_opt=f_x_1
    while (abs(x_max-x_min)>precision):
        if (f_x_1<=f_x_2):
            x_max=x_2
            x_1,x_2=x_min+((3-math.sqrt(5))/2)*(x_max-x_min), x_1
            f_x_1,f_x_2=abs_x(x_1),f_x_1
            counter+=1
            x_opt=x_1
            f_opt=f_x_1
        else:
            x_min=x_1
            x_1,x_2=x_2,x_max+((math.sqrt(5)-3)/2)*(x_max-x_min)
            f_x_1,f_x_2=f_x_2, abs_x(x_2)
            counter+=1
            x_opt=x_2
            f_opt=f_x_2
    return({'x_opimal':x_opt, 'minimum':f_opt, 'counter':counter})

def goldratio_sin_x(x_min=0.1, x_max=1):
    x_1=x_min+((3-math.sqrt(5))/2)*(x_max-x_min)
    x_2=x_max+((math.sqrt(5)-3)/2)*(x_max-x_min)
    f_x_1=sin_x(x_1)
    f_x_2=sin_x(x_2)
    counter=2
    x_opt=x_1
    f_opt=f_x_1
    while (abs(x_max-x_min)>precision):
        if (f_x_1<=f_x_2):
            x_max=x_2
            x_1,x_2=x_min+((3-math.sqrt(5))/2)*(x_max-x_min), x_1
            f_x_1,f_x_2=sin_x(x_1),f_x_1
            counter+=1
            x_opt=x_1
            f_opt=f_x_1
        else:
            x_min=x_1
            x_1,x_2=x_2,x_max+((math.sqrt(5)-3)/2)*(x_max-x_min)
            f_x_1,f_x_2=f_x_2, sin_x(x_2)
            counter+=1
            x_opt=x_2
            f_opt=f_x_2
    return({'x_opimal':x_opt, 'minimum':f_opt, 'counter':counter})
                         
print(brute_force_x_exp_3 (0, 1))
print(brute_force_abs_x (0, 1))
print(brute_force_sin_x (0.1, 1))
print()
print(dichotomy_x_exp_3 (0, 1))
print(dichotomy_abs_x (0, 1))
print(dichotomy_sin_x (0.1, 1))
print()
print(goldratio_x_exp_3 (0, 1))
print(goldratio_abs_x (0, 1))
print(goldratio_sin_x (0.1, 1))
```
