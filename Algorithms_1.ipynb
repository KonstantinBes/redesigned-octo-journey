```python
%matplotlib inline

import os
import numpy as np
import matplotlib.pyplot as plt
import pylab
import random
import math
import time

def create_vect(vect_size):
    vect = [random.randint(1, vect_size) for i in range(vect_size)]
    return(vect)
def const_funct(vect):
    return(vect[0])
def product_funct(vect):
    product = 1
    for i in vect:
        product*=i
    return(product)
def euclidean_funct(vect):
    return(math.sqrt(sum(i**2 for i in vect)))
def polynom_funct(vect):
    x=1.5
    poly_value=0
    for i in range(len(vect)):
        poly_value+=vect[i]*x**i
    return(poly_value)
def horner_funct(vect):
    x=1.5
    poly_value=0
    for i in range(len(vect)-1, -1, -1):
        poly_value=vect[i]+x*poly_value
    return(poly_value)
def create_matrix(vect_size):
    matrix = [create_vect(vect_size) for i in range(vect_size)]
    return(matrix)
def matrix_production(matrix):
    matrix_prod=[]
    for row in range(len(matrix)):
        product_row=[]
        for element in range(len(matrix[row])):        
            product_row+=[sum(matrix[row][i]*matrix[i][element] for i in range(len(matrix)))]
        matrix_prod+=[product_row]
    return(matrix_prod)
def bubble_sort(vect):
    for i in range(len(vect)-1):
        already_sorted = True
        for j in range(len(vect)-1-i):
            if(vect[j]>vect[j+1]):
                vect[j],vect[j+1]=vect[j+1],vect[j]
                already_sorted = False
        if(already_sorted==True):
            break
    return(vect)

def output_const (vect_sizes, result_table):
    x = vect_sizes
    y = result_table
    fig, ax = plt.subplots(1, 1, figsize=(8,6))
    ax.grid()
    ax.set_xlabel('vect_size, (units)')
    ax.set_ylabel('time, s')
    pylab.xlim(0,x[len(vect_sizes)-1])
    pylab.ylim(0,1.2e-6)
    line1, = ax.plot(x, y['const'], linewidth=2, label='Practic')
    line2, = ax.plot(x, [y['const'][len(vect_sizes)-1] for i in range(len(vect_sizes))], dashes=[10, 5, 10, 5],
                     linewidth=2, label='Theory')
    ax.legend(loc='lower right', borderaxespad=0.1)
    fig.savefig('A:\Information\Kostya\Master\Const.png')
    plt.close(fig)
def output_summ (vect_sizes, result_table):
    x = vect_sizes
    y = result_table
    fig, ax = plt.subplots(1, 1, figsize=(8,6))
    ax.grid()
    ax.set_xlabel('vect_size, (units)')
    ax.set_ylabel('time, s')
    pylab.xlim(0,x[len(vect_sizes)-1])
    pylab.ylim(0,5e-6)
    line1, = ax.plot(x, y['summ'], linewidth=2, label='Practic')
    line2, = ax.plot(x, [y['summ'][len(vect_sizes)-1]/(vect_sizes[len(vect_sizes)-1])*i for i in vect_sizes],
                     dashes=[10, 5, 10, 5], linewidth=2, label='Theory')
    ax.legend(loc='upper left', borderaxespad=0.1)
    fig.savefig('A:\Information\Kostya\Master\Summ.png')
    plt.close(fig)
def output_product (vect_sizes, result_table):
    x = vect_sizes
    y = result_table
    fig, ax = plt.subplots(1, 1, figsize=(8,6))
    ax.grid()
    ax.set_xlabel('vect_size, (units)')
    ax.set_ylabel('time, s')
    pylab.xlim(0,x[len(vect_sizes)-1])
    pylab.ylim(0,5e-5)
    line1, = ax.plot(x, y['product'], linewidth=2, label='Practic')
    line2, = ax.plot(x, [y['product'][len(vect_sizes)-1]/(vect_sizes[len(vect_sizes)-1])*i for i in vect_sizes],
                     dashes=[10, 5, 10, 5], linewidth=2, label='Theory')
    ax.legend(loc='upper left', borderaxespad=0.1)
    fig.savefig('A:\Information\Kostya\Master\Product.png')
    plt.close(fig)
def output_euclidean (vect_sizes, result_table):
    x = vect_sizes
    y = result_table
    fig, ax = plt.subplots(1, 1, figsize=(8,6))
    ax.grid()
    ax.set_xlabel('vect_size, (units)')
    ax.set_ylabel('time, s')
    pylab.xlim(0,x[len(vect_sizes)-1])
    pylab.ylim(0,1.5e-4)
    line1, = ax.plot(x, y['euclidean'], linewidth=2, label='Practic')
    line2, = ax.plot(x, [y['euclidean'][len(vect_sizes)-1]/(vect_sizes[len(vect_sizes)-1])*i for i in vect_sizes],
                     dashes=[10, 5, 10, 5], linewidth=2, label='Theory')
    ax.legend(loc='upper left', borderaxespad=0.1)
    fig.savefig('A:\Information\Kostya\Master\Euclidean.png')
    plt.close(fig)    
def output_polynom (vect_sizes, result_table):
    x = vect_sizes
    y = result_table
    fig, ax = plt.subplots(1, 1, figsize=(8,6))
    ax.grid()
    ax.set_xlabel('vect_size, (units)')
    ax.set_ylabel('time, s')
    pylab.xlim(0,x[len(vect_sizes)-1])
    pylab.ylim(0,10e-5)
    line1, = ax.plot(x, y['polynom'], linewidth=2, label='Practic')
    line2, = ax.plot(x, [y['polynom'][len(vect_sizes)-1]/(vect_sizes[len(vect_sizes)-1])*i for i in vect_sizes],
                     dashes=[10, 5, 10, 5], linewidth=2, label='Theory')
    ax.legend(loc='upper left', borderaxespad=0.1)
    fig.savefig('A:\Information\Kostya\Master\Polynom.png')
    plt.close(fig)    
def output_horner (vect_sizes, result_table):
    x = vect_sizes
    y = result_table
    fig, ax = plt.subplots(1, 1, figsize=(8,6))
    ax.grid()
    ax.set_xlabel('vect_size, (units)')
    ax.set_ylabel('time, s')
    pylab.xlim(0,x[len(vect_sizes)-1])
    pylab.ylim(0,4e-5)
    line1, = ax.plot(x, y['horner'], linewidth=2, label='Practic')
    line2, = ax.plot(x, [y['horner'][len(vect_sizes)-1]/(vect_sizes[len(vect_sizes)-1])*i for i in vect_sizes],
                     dashes=[10, 5, 10, 5], linewidth=2, label='Theory')
    ax.legend(loc='upper left', borderaxespad=0.1)
    fig.savefig('A:\Information\Kostya\Master\Horner.png')
    plt.close(fig)    
def output_bubble (vect_sizes, result_table):
    x = vect_sizes
    y = result_table
    fig, ax = plt.subplots(1, 1, figsize=(8,6))
    #ax.set(title = 'Bubble', xticks=[], yticks=[])
    ax.grid()
    ax.set_xlabel('vect_size, (units)')
    ax.set_ylabel('time, s')
    pylab.xlim(0,x[len(vect_sizes)-1])
    pylab.ylim(0,0.0013)
    line1, = ax.plot(x, y['bubble'], linewidth=2, label='Practic')
    line2, = ax.plot(x, [y['bubble'][len(vect_sizes)-1]/((vect_sizes[len(vect_sizes)-1])**2)*i**2 for i in vect_sizes],
                     dashes=[10, 5, 10, 5], linewidth=2, label='Theory')
    ax.legend(loc='upper left', borderaxespad=0.1)
    fig.savefig('A:\Information\Kostya\Master\Bubble.png')
    plt.close(fig)
def output_matrix (vect_sizes, result_table):
    x = vect_sizes
    y = result_table
    fig, ax = plt.subplots(1, 1, figsize=(8,6))
    ax.grid()
    ax.set_xlabel('vect_size, (units)')
    ax.set_ylabel('time, s')
    pylab.ylim(0,9)
    line1, = ax.plot(x, y['matrix_production'], linewidth=2, label='Practic')
    line2, = ax.plot(x, [y['matrix_production'][len(vect_sizes)-1]/(vect_sizes[len(vect_sizes)-1]**3)*i**3 for i in vect_sizes],
                     dashes=[10, 5, 10, 5], linewidth=2, label='Theory')
    ax.legend(loc='lower right', borderaxespad=0.1)
    fig.savefig('A:\Information\Kostya\Master\Matrix.png')
    plt.close(fig)


"""def output_graph(vect_sizes, result_table):
    x = vect_sizes
    y = result_table
    fig, ax = plt.subplots(1, 1, figsize=(15,9))
    ax.grid()
    ax.set_xlabel('vect_size, (units)')
    ax.set_ylabel('time, (s*10**(-7))')
    pylab.xlim(0,x[len(vect_sizes)-1])
    pylab.ylim(0,8*10**-7)
    line1, = ax.plot(x, y['const'], linewidth=2,
                     label='const')
    line2, = ax.plot(x, y['summ'], dashes=[10, 5, 10, 5], linewidth=2,
                     label='summ')
    line3, = ax.plot(x, y['product'], dashes = [30, 5, 5, 5], linewidth=2,
                     label='product')
    line4, = ax.plot(x, y['euclidean'], dashes=[10, 5, 10, 5], linewidth=2,
                     label='euclidean')
    line5, = ax.plot(x, y['polynom'], linewidth=2,
                     label='polynom')
    line6, = ax.plot(x, y['horner'], dashes=[30, 5, 30, 5], linewidth=2,
                     label='horner')
    line7, = ax.plot(x, y['bubble'], linewidth=2,
                     label='bubble')
    line8, = ax.plot(x, y['matrix_production'], linewidth=2,
                     label='matrix_production')
    ax.legend(loc='upper left', borderaxespad=0.0)
    fig.savefig('A:\Information\Kostya\Master\Plot.png')
    plt.close(fig)
def theoretical_graph(vect_sizes):
    x = vect_sizes
    y = vect_sizes
    fig, ax = plt.subplots(1, 1, figsize=(15,9))
    ax.grid()
    ax.set_xlabel('vect_size, (units)')
    ax.set_ylabel('time, (operations)')
    pylab.xlim(0,x[len(vect_sizes)-1])
    pylab.ylim(0,x[len(vect_sizes)-1]*math.log(len(vect_sizes)))
    line1, = ax.plot(x, [y[0] for i in range(len(vect_sizes))], linewidth=2,
                     label='const')
    line2, = ax.plot(x, y, dashes=[10, 5, 10, 5], linewidth=2,
                     label='summ')
    line3, = ax.plot(x, y, dashes = [30, 5, 5, 5], linewidth=2,
                     label='product')
    line4, = ax.plot(x, [i*2 for i in y], dashes=[10, 5, 10, 5], linewidth=2,
                     label='euclidean')
    line5, = ax.plot(x, [i*math.log(i) for i in y], linewidth=2,
                     label='polynom')
    line6, = ax.plot(x, [i*2 for i in y], dashes=[30, 5, 30, 5], linewidth=2,
                     label='horner')
    line7, = ax.plot(x, [i**2 for i in y], linewidth=2,
                     label='bubble')
    line8, = ax.plot(x, [i**3 for i in y], linewidth=2,
                     label='matrix_production')
    ax.legend(loc='lower right', borderaxespad=0.0)
    fig.savefig('A:\Information\Kostya\Master\Theoretiacal_plot.png')
    plt.close(fig)"""
def create_result_table(vect_sizes, number_of_tests):
    functions = {'const': [], 'summ': [], 'product': [], 'euclidean': [], 'polynom': [], 'horner': [], 'bubble': [], 'matrix_production': []}
    result_table = {'const': [], 'summ': [], 'product': [], 'euclidean': [], 'polynom': [], 'horner': [], 'bubble': [], 'matrix_production': []}
    before=time.perf_counter()
    after=before
    for vect_size in vect_sizes:
        vect=create_vect(vect_size)
        matrix=create_matrix(vect_size)
        for test_case in range(number_of_tests-1):    # Measurement of executabale time for each function
            before=time.perf_counter()
            const_funct(vect)
            after=time.perf_counter()
            functions['const'].append(after-before)
            before=time.perf_counter()
            sum(vect)
            after=time.perf_counter()
            functions['summ'].append(after-before)
            before=time.perf_counter()
            product_funct(vect)
            after=time.perf_counter()
            functions['product'].append(after-before)
            before=time.perf_counter()
            euclidean_funct(vect)
            after=time.perf_counter()
            functions['euclidean'].append(after-before)
            before=time.perf_counter()
            polynom_funct(vect)
            after=time.perf_counter()
            functions['polynom'].append(after-before)
            before=time.perf_counter()
            horner_funct(vect)
            after=time.perf_counter()
            functions['horner'].append(after-before)    
            before=time.perf_counter()
            bubble_sort(vect)
            after=time.perf_counter()
            functions['bubble'].append(after-before)
            before=time.perf_counter()
            matrix_production(matrix)
            after=time.perf_counter()
            functions['matrix_production'].append(after-before)
        for key in result_table.keys():
            result_table[key].append(sum(functions[key])/len(functions[key]))
    return(result_table)
vect_sizes = [i for i in range(1, 502, 50)]
number_of_tests = 10
result_table = create_result_table(vect_sizes, number_of_tests)
result_table = {'const': [9.505555580189038e-07, 9.005555580189038e-07, 9.50555580189038e-07, 8.505555580189038e-07, 9.505555580189038e-07, 8.505555580189038e-07, 10.505555580189038e-07, 9.323930561945923e-07, 9.505555580189038e-07, 9.00555580189038e-07, 9.05555580189038e-07], 'summ': [4.179999753735804e-07, 1.0453333692971178e-06, 1.4064074060081988e-06, 1.7770277812208708e-06, 2.098222244563254e-06, 2.5847592766055e-06, 3.00288889948639e-06, 3.340250006860717e-06, 3.7504567916147168e-06, 4.219255551005416e-06, 4.661565655091309e-06], 'product': [5.701111452556992e-07, 3.725111128005665e-06, 6.930703724113603e-06, 1.0671722242881919e-05, 1.492328889677689e-05, 1.996235185422982e-05, 2.525049205995322e-05, 3.0152652780341465e-05, 3.5806864206627066e-05, 4.159984445222007e-05, 4.790838384547952e-05], 'euclidean': [2.357000034357447e-06, 1.7751444425408004e-05, 3.0903370400294616e-05, 4.4188361155445615e-05, 5.7830755607331716e-05, 7.174681485998412e-05, 8.499928575445896e-05, 9.819812503843018e-05, 0.00011162729632264983, 0.00012526712224724987, 0.00013880096972258404], 'polynom': [1.596444488111754e-06, 1.0681388883110938e-05, 1.8980518503937032e-05, 2.6674666641055308e-05, 3.410402217317217e-05, 4.1451648106279514e-05, 4.915444441437666e-05, 5.691284719154687e-05, 6.475485181987478e-05, 7.292519996756002e-05, 8.107868683914098e-05], 'horner': [1.2164445150928159e-06, 4.7323333850120735e-06, 7.969703734421205e-06, 1.079519446041862e-05, 1.3516800016885907e-05, 1.6237185204941112e-05, 1.917404764147553e-05, 2.2174944471468432e-05, 2.5349469165065122e-05, 2.854668890298247e-05, 3.1746575775351365e-05], 'bubble': [8.742221729739363e-07, 2.2388722199037955e-05, 6.411281480756804e-05, 0.0001230529444329578, 0.00020653215555689207, 0.00031965846297107755, 0.0004486562222284972, 0.0006152271111103699, 0.0007963892963002932, 0.0010062276333428599, 0.001229930707079738], 'matrix_production': [3.535111116232454e-06, 0.017708440888908526, 0.09666317159260668, 0.2846890928055548, 0.6256018404222232, 1.1587598123333347, 1.940821779682542, 3.0182057227638857, 4.4551262463456665, 6.287779554866661, 8.596413447474747]}
output_const(vect_sizes, result_table)
output_summ(vect_sizes, result_table)
output_product(vect_sizes, result_table)
output_euclidean(vect_sizes, result_table)
output_polynom(vect_sizes, result_table)
output_horner(vect_sizes, result_table)
output_bubble(vect_sizes, result_table)
output_matrix(vect_sizes, result_table)
print (result_table)
```
